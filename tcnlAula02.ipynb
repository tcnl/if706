```python
import numpy as np
import pandas as pd
```

```python
#Printar msgs
print("Dale dale dele doli")

#Variaveis dinam tipadas
nome = "Thiago"
idade = 20
peso = 87.5

print(nome, idade, peso)
```

```python
#Listas (não arrays) podem ter valores diversos
dados = [nome, idade, peso, 13]
print(dados)

#Python ainda tem tuplas ordenadas
ponto = (19, 11, -1)
print(ponto)

#Dicionario tb kappa
dic = {'nome':nome, 'idade':idade, 'peso':peso, 'end':'Rua Herminio Alves de Queiros, 650'}
print(dic['end'])
```

```python
#Condicionais sao feitos da seguinte forma
if idade > 30:
    #Python usa identacao pra definir o nivel/escopo das instrucoes condicionais
    #cuidado pra na fazer caca
    print(nome,"já quase na meia-vida")
    print("Erro!")
else:
    print(nome,"ainda ta jovem")
```

```python
#Repeticoes podem ser feitas de duas formas
#Um - usando lacos while
vezes = 1
while vezes <= 5:
    teste = vezes*2
    print("Já passei por aqui {0} vez(es) e vezes x 2 = {1}".format(vezes, teste))
    vezes += 1
```

```python
#Dois - enumerando os elementos de uma lista (objeto interavel)
for vezes in [1, 3, 5, 7, 9]:
    teste = 2*vezes
    print("Ja passei por aqui {0} vez(es) e vezes x 2 = {1}".format(vezes, teste))
print("========================")
#Ou entao
for vezes in range(1,6,1):
    #form in range(inicio, atéExclusivo, incremento)
    print("Já sabem que eu passei por aqui {0} vez(es)".format(vezes))
```

```python
#Python possui um recurso massa pra gerar listas de elementos ja populadas
#Eh tipo assim, cria a variavel com um 'laco' dentro que ja popula
#Nome chique: list comprehension

quadrados = [a**2 for a in range(1,6)]
paresQuadrado = [a**2 for a in range(1,20) if a%2==0]

print(quadrados)
print(paresQuadrado)

#Pode fazer uns bagulho mais doido ainda
primPotencias = [[base**expo for expo in range(4)] for base in range(4)]
print(primPotencias)
```

```python
#É assim que se define funcoes
def quadrado(x):
    return x**2

print(quadrado(3))

#funcoes mais simples podem ser definidas como funcoes lambda lambda lambda nerds
f = lambda x: x**2
print(f(4))
```

```python
#Algumas outras coisas legais e uteis
#map = aplica uma funcao a cada elemento de uma lista e retorna o resultado
a = map(f, range(6))
print(a, list(a))

#zip = combina elementos de varios iteraveis em tuplas
a = zip(range(6), map(f, range(6)))
print(a,list(a))

#filter = retorna os elementos de um iteravel para os quais o resultado de uma funcao e verdadeiro
a = filter(lambda x: x%2==0, range(10))
print(a,list(a))
```

```python
#Agora sim a gente vai criar 'arrays' propriamente ditos, ajuda a integrar com C/C++ e Fortran
arr1 = np.array([1,2,3,4,5])
print(type(arr1), arr1)

#Bidimensional
arr2 = np.array([range(5),range(5,10)])
print(arr2)

#Os tamanhos das dimensoes podem ser obtidos com o metdo shape
print(arr1.shape, arr2.shape)

#Pode-se alterar a forma de um vetor mesmo apos a sua criacao
arr2.shape = (2,5)
print(arr2)
```

```python
#NumPy faz operacoes vetorizadas... d o p e
a = np.arange(50,step=5,dtype=np.float)
#form .arrange(kMax, incremento, tipo)
print(a)
arr1 = a.reshape((2,5))
print(arr1)
arr2 = np.linspace(.1,1,num=10,dtype=np.float).reshape((2,5))
print(arr2,"\n=========")
print(arr1 - arr2,"\n=======")
print(arr2*arr1,"\n========")
print(arr1*arr2,"\n========")
print(np.dot(arr2,arr1.T))
```

```python
#Para acessar um elemento, basta informar seu indice em cada eixo
# array(linha, coluna)
print(arr1[1,4])

#Podemos tambem selecionar uma porcao dos elementos (slicing)
aslice = arr1[:, 2:4]
#formato arr1[todas as linhas, colunas 2 e 3]
print(aslice)

#Os resultados de um slice (basico) é uma visao do vetor original
#Ou seja, o vetor pode ser alterado atraves do slice
aslice[:,:] = 0
print(arr1)
```

```python
#NumPy oferece ainda mecanismos para selecao avancada de porcoes
#Mas isso buga aquela ideia do slice, entao se mexer, nao muda o array original
#Usando uma lista de indices
aslice2 = arr1[:,[0, 2, 4]]
print(aslice2)
aslice2[:,:] = 0
print(arr1)

#Ou mascaras booleanas
print(arr1[arr1>10])
```
