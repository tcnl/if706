```python
import numpy as np
import scipy as sp
import pandas as pd

import re

from IPython.display import display
```

```python
usuarios = pd.read_csv(
    "http://files.grouplens.org/datasets/movielens/ml-100k/u.user",
    sep='|',header=None, names=["user_id", "age", "gender", "occupation", "zip_code"])
filmes = pd.read_csv(
    "http://files.grouplens.org/datasets/movielens/ml-100k/u.item",
    sep='|',header=None, names=["movie_id", "movie_title",  "release_date", "video_release_date", "IMDb_URL", 
                                "unknown", "Action", "Adventure", "Animation", 
        "Children", "Comedy", "Crime", "Documentary", "Drama", "Fantasy","FilmNoir", "Horror", 
                                "Musical", "Mystery", "Romance", "SciFi","Thriller", "War", "Western"],
encoding='latin1')
avaliacoes = pd.read_csv(
    "http://files.grouplens.org/datasets/movielens/ml-100k/u.data",
    sep='\t',header=None, names=["user_id", "movie_id", "rating", "timestamp"])
```

```python
filmes['release_year'] = filmes.release_date.apply(
lambda x: not x is np.nan and
re.search("\d+\-\w+\-(\d+)",str(x)).group(1) or None)
filmes[['movie_title', 'release_date', 'release_year']].head(15)
```

```python
#Pandas tem funcoes estatisticas prontas, como a apply, usada anteriormente, que aplica uma funcao a todos os elementos
print("Mediana de idade dos usuarios ", usuarios.age.median())

#pior avaliacao
piorAvaliacao = avaliacoes.rating.values.argmin()
print(filmes[filmes.movie_id == avaliacoes.movie_id.iloc[piorAvaliacao]].movie_title.iloc[0])

avaliacoes.rating.value_counts()
```

```python
#No in anterior vimos que da pra fazer cruzamento de tabelas
#Assim como em banco de dados (SQL)
#Advinha, pandas faz isso. G O D

#A forma mais simples de juntar diferentes dados e contenacao
A = pd.Series(["A{}".format(a) for a in range(4)],
             index=range(4), name="A")
B = pd.Series(["A{}".format(b) for b in range(4)],
             index=range(4), name="B")
C = pd.Series(["C{}".format(c) for c in range(5)],
             index=range(5), name="C")
pd.concat([A,B,C],axis=1)
```

```python
#Pra quem curte SQL, tem outras funcoes como merge
#A Sintaxe é como segue:
# pd.merge(left, right, how='inner', 
# on=None, left_on=None, right_on=None,
# left_index=False, right_index=False, sort=True,
# suffixes=('_x', '_y'),copy=True,indicator=False)

avaliacoesFilmes = pd.merge(avaliacoes,
                          filmes[["movie_id","movie_title",
                                 "release_year"]],
                          on="movie_id")
avaliacoesFilmes.tail(15)
```

```python
avaliacoesFilmes.release_year.fillna(value=-1, inplace=True)
avaliacoesFilmes['release_year'] = \
avaliacoesFilmes.release_year.astype(np.int)
avalFilmUsu = pd.merge(avaliacoesFilmes, usuarios[["user_id","occupation"]],
                      on="user_id")
selecao = avalFilmUsu[(1980 <= avalFilmUsu.release_year) &
                     (avalFilmUsu['release_year'] < 1990) &
                     (avalFilmUsu.occupation == 'programmer')]
selecao.rating.mean(), selecao.rating.std()
```

```python
#Outro recurso bolado do Pandas é o group-by
#Podemos agrupar dados a partir de um criterio para aplicar determinada funcao
#Por ex, pode ser util saber a media das avaliações por genero do filme, ou faixa de idade

#Faz assim oh:
# 1. Split: divide os dados conforme criterio
# 2. Apply: aplica uma determinada funcao a cada um dos grupos
# 3. Combine: reunir resultados em uma unica estrutura de dados

#A funcao que sera aplicada aos grupos pode ter como objetivo:
# Sumarizar informacoes: computar estatisticas descritivas dos grupos (media, desvios, minimo/maximo, contar valores, etc)
# Transformar valores: normalizar dados nos grupos: gerar novos atributos; prencher dados ausentes
# Filtrar dados: eliminar grupos a partir de estatisticas computadas para o grupo

#=========EXEMPLOS=========#
usuariosPorProfissao = usuarios.groupby(usuarios.occupation,
                                       as_index=True)
usuariosPorProfissao.mean()
```

```python
print(filmes.release_year.dropna().astype(np.int).describe())
#print(pd.cut(filmes.release_year.dropna().astype(np.int),bins=10).head(15))
decada = pd.cut(filmes.release_year.fillna(value=0).astype(np.int),
               bins=[1900,1940,1950,1960,1970,1980,1990,2000],
               labels=['{}s'.format(a) for a in np.arange(30, 100, 10)])
filmePorDecada = filmes.groupby(decada)
filmePorDecada[filmes.columns[5:]].sum()
```

```python
b = filmePorDecada[filmes.columns[5:]].sum()
sizes = filmePorDecada.size()
print(sizes)
b.T/sizes
```
